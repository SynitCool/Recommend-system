```python
def unique_value(feature):
    unique = []
    for val in df[feature]:
        if ',' in val:
            ftr = val.split(',')
            unique.extend(ftr)
        else:
            unique.append(val)
    
    unique = pd.Series(unique)
    unique = list(unique.apply(lambda x: x.strip()))
    
    ftr, counts = np.unique(unique, return_counts=True)

    counts = pd.DataFrame(data={'ftr':ftr,
                                'counts':counts}).sort_values('counts', ascending=False)

    print(f"There are {counts['ftr'].shape[0]} {feature}")
```

```python
unique_value('cast')
unique_value('country')
unique_value('listed_in')
unique_value('director')
```

```python
def unique_value_cl(feature, rng_cl):
    unique = []

    value = df[feature].str.split(',', expand=True).loc[:, :rng_cl]
    value = value.fillna('missing')
    for column in value.columns:
        value[column].apply(lambda x: unique.append(x))

    unique = pd.Series(unique)
    unique = list(unique.apply(lambda x: x.strip()))
    
    val, counts = np.unique(unique, return_counts=True)

    counts = pd.DataFrame(data={'value':val,
                                'counts':counts}).sort_values('counts', ascending=False)

    print(f"There are {counts['value'].shape[0]} {feature}")
```

```python
unique_value_cl('cast', 4)
unique_value_cl('country', 1)
unique_value_cl('director', 1)
unique_value_cl('listed_in', 1)
```

```python
def null_percent_cl(feature, rng_cl):
    ftr = df[feature].str.split(',', expand=True).loc[:, :rng_cl]
    for column in ftr.columns:
        thres = ftr.loc[:, column].isnull().sum()
        print(f"There are {round((thres/ftr.shape[0]) * 100, 2)} null percent in {feature} column {column}")
    print('\n-------')
```

```python
null_percent_cl('cast', 3)
null_percent_cl('country', 0)
null_percent_cl('director', 0)
null_percent_cl('listed_in', 0)
```

```python
unique_value_cl('cast', 3)
unique_value_cl('country', 0)
unique_value_cl('director', 0)
unique_value_cl('listed_in', 0)
```

```python
def transformed_cl(feature, rng_cl):
    unique = []

    ftr = df[feature].str.split(',', expand=True).loc[:, :rng_cl]
    ftr = ftr.fillna('missing')
    for column in ftr.columns:
        ftr[column].apply(lambda x: unique.append(x))

    unique = pd.Series(unique)
    unique = list(unique.apply(lambda x: x.strip()))
    
    unique = np.unique(unique)
    ftr = pd.DataFrame(columns=unique)
    
    ind = []
    for val in df[feature]:
        trans = []
        for column in ftr.columns:
            if column in val:
                trans.append(1)
            else:
                trans.append(0)
        ind.append(trans)
        
    columns = []
    for unq in unique:
        columns.append(f"{unq}_{feature}")
    
    return pd.DataFrame(ind, columns=columns)
```

```python
def transformed(feature):
    unique = df[feature].value_counts().index
    ftr = pd.DataFrame(columns=unique)
    
    ind = []
    for val in df[feature]:
        trans = []
        for column in ftr.columns:
            if column == val:
                trans.append(1)
            else:
                trans.append(0)
        ind.append(trans)
        
    columns = []
    for unq in unique:
        columns.append(f"{unq}_{feature}")
        
    return pd.DataFrame(ind, columns=columns)
```

```python
start = time.time()

casts = transformed_cl('cast', 3).reset_index(drop=True)
directors = transformed_cl('director', 0).reset_index(drop=True)
countries = transformed_cl('country', 0).reset_index(drop=True)
genres = transformed_cl('listed_in', 0).reset_index(drop=True)

types = transformed('type').reset_index(drop=True)
ratings = transformed('rating').reset_index(drop=True)

print(f'Finish : {time.time() - start}')
```

```python
df = df[['title', 'release_year']].reset_index(drop=True)

df = pd.concat([df, casts, directors, genres, ratings, countries, types], axis=1)

df.shape
```

```python
ori_df = ori_df.dropna()
ori_df = ori_df.reset_index(drop=True)
```

```python
X = df.drop(['title'], axis=1)
y = df['title']
```

```python
neigh = NearestNeighbors(algorithm='brute')

neigh.fit(X)
```

```python
def find_movies(title):
    if title in df['title'].values:
        x = df[df['title'] == title].drop(columns=['title'])
        distances, movies = neigh.kneighbors(x, 6, return_distance=True)
        
        
        result = ori_df.copy()
        result = result.iloc[movies[0]]
        result['distance'] = distances[0]
        
        return result
    else:
        print("Title is not in dataset")
```
