```python
import pandas as pd
import numpy as np

from sklearn.neighbors import KNeighborsClassifier, NearestNeighbors
```

```python
df = pd.read_csv('Dataset\\engineered_title_netflix_dataset.csv')
ori_df = pd.read_csv('Dataset\\cleaned_title_netflix_dataset.csv')

df.head()
```

```python
ori_df.info()
```

```python
ori_df = ori_df.dropna()
ori_df = ori_df.reset_index(drop=True)
```

# Fitting to Model

```python
X = df.drop(['title'], axis=1)
y = df['title']
```

```python
neigh = NearestNeighbors(n_neighbors=10, algorithm='brute')

neigh.fit(X)
```

```python
def find_movies(title):
    if title in df['title'].values:
        x = df[df['title'] == title].drop(columns=['title'])
        distances, movies = neigh.kneighbors(x, 6, return_distance=True)
        
        
        result = ori_df.copy()
        result = result.iloc[movies[0]]
        result['distance'] = distances[0]
        
        return result
    else:
        print("Title is not in dataset")
```
